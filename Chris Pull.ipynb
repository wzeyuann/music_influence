```python
import numpy as np
import pandas as pd
```

```python
artists = pd.read_csv('data/allmusic/artists_cleaned.csv')
influences = pd.read_csv('data/allmusic/influences_cleaned.csv')
```

```python
artists.head()
```

```python
influences.head()
```

```python
influencer_merged = pd.merge(influences, artists, left_on='influencer_id', right_on='id')
influencer_merged_subset = influencer_merged[['influencer_name', 'follower_name', 'influencer_id', 'follower_id', 'main_genre', 'active_start']]
influencer_merged_subset.columns = ['influencer_name', 'follower_name', 'influencer_id', 'follower_id', 'influencer_main_genre', 'influencer_active_start']
```

```python
follower_merged = pd.merge(influencer_merged_subset, artists, left_on='follower_id', right_on='id')
follower_merged_subset = follower_merged[['influencer_name', 'follower_name', 'influencer_id', 'follower_id', 'influencer_main_genre', 'influencer_active_start', 'main_genre', 'active_start']]
follower_merged_subset.columns = ['influencer_name', 'follower_name', 'influencer_id', 'follower_id', 'influencer_main_genre', 'influencer_active_start', 'follower_main_genre', 'follower_active_start']
```

```python
follower_merged_subset[follower_merged_subset['follower_name'] == 'Charlie Parker']
```

```python
follower_merged_subset.to_csv('data/influence_data_for_chris.csv', index=False)
```
